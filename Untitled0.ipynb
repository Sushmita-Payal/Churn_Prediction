**Many of the columns are yes, no etc. Let's print unique values in object columns to see data values**

```python
for column in df:
  if df[column].dtypes=='object':
     print(f'{column} : {df[column].unique()}')
```

```python
def print_unique_col_values(df):
    for column in df:
        if df[column].dtypes=='object':
            print(f'{column} : {df[column].unique()}')
```

```python
print_unique_col_values(df01)
```

**Some of the columns have no internet service or no phone service, replacing it with simple No**

```python
df01.replace('No internet service','No',inplace=True)
df01.replace('No phone service','No',inplace=True)
print_unique_col_values(df01)
```

**Converting Yes and No to Binary**

```python
yes_no_columns = ['Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'OnlineSecurity', 'OnlineBackup',
                  'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling','Churn']


for col in yes_no_columns:
  df01[col].replace({'Yes': 1, 'No': 0}, inplace=True)
```

```python
for col in df01:
  print(f'{col}: {df01[col].unique()}')
```

```python
df01['gender'].replace({'Female': 1, 'Male': 0}, inplace=True)
df01['gender'].unique()
```

**One hot encoding for categorical columns**

```python
df2=pd.get_dummies(data=df01,columns=['InternetService','Contract','PaymentMethod'])
df2.columns
```

```python
df2.sample(5)
```

```python
df2.dtypes
```

```python
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

```python
for col in df2:
    print(f'{col}: {df2[col].unique()}')
```

**Train and Test split**

```python
X = df2.drop('Churn',axis='columns')
y = df2['Churn']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train[:10]
```

```python
len(X_train.columns)
```

**Building a model (ANN) in tensorflow/keras**

```python
import tensorflow as tf
from tensorflow import keras


model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(26,), activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

# opt = keras.optimizers.Adam(learning_rate=0.01)

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test, y_test)
```

```python
yp = model.predict(X_test)
yp[:5]
```

```python
y_test[:5]
```

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test,y_pred))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
